```python
def print_stat(train_df, test_df, train_subset_df=None, valid_subset_df=None):
    y_cum = [c for c in train_df.columns if 'y_cum_' in c]

    print('Statistics')
    print('-'*20)
    print('train_df: ', len(train_df))
    if train_subset_df!=None:
        print('train_subset_df: ', len(train_subset_df))
        print('valid_subset_df: ', len(valid_subset_df))
    print('test_df: ', len(test_df))

    print('-'*20)
    y_cum_train = train_df[y_cum].replace({-1:0, 2:0}).sum(axis=1).values
    if train_subset_df!=None:
        y_cum_train_subset = train_subset_df[y_cum].replace({-1:0, 2:0}).sum(axis=1).values
        y_cum_valid_subset = valid_subset_df[y_cum].replace({-1:0, 2:0}).sum(axis=1).values
    y_cum_test = test_df[y_cum].replace({-1:0, 2:0}).sum(axis=1).values

    print("train:\t",y_cum_train.sum()/len(y_cum_train))
    if train_subset_df!=None:
        print("train_subset:\t", y_cum_train_subset.sum()/len(y_cum_train_subset))
        print("valid_subset:\t", y_cum_valid_subset.sum()/len(y_cum_valid_subset))
    print("test:\t", y_cum_test.sum()/len(y_cum_test))
    print('-'*20)
```
